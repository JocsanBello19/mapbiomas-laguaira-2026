```python
import pandas as pd
import numpy as np
import os

# Nombre del archivo original subido
archivo_entrada = "Serie temporal de Cobertura • Cobertura por clase • 1985 - 2024.csv"
archivo_salida = "Metricas_Criticas_LaGuaira_1985_2024.csv"

if not os.path.exists(archivo_entrada):
    raise FileNotFoundError(f"⚠️ El archivo '{archivo_entrada}' no se encuentra en la carpeta actual. Verifícalo.")

print("⏳ Procesando datos de MapBiomas Venezuela Colección 3...")

# 1. Cargar el archivo manejando la codificación UTF-8 estándar
df = pd.read_csv(archivo_entrada, encoding='utf-8')

# 2. Definir los años hito del estudio
anos_clave = ['1985', '1999', '2024']

# 3. Filtrar las clases críticas del Nivel 2 para La Guaira
clases_interes = ["Bosque", "Arbustal", "Sabana/Herbazal", "Área urbanizada", "Manglar"]
df_filtrado = df[df['Nivel 2'].isin(clases_interes)].copy()

# 4. Estructurar la tabla pivote con los años de interés
tabla_resumen = df_filtrado[['Nivel 2'] + anos_clave].set_index('Nivel 2')

# 5. Aplicar Control Métrico Estricto (Cálculos robustos)
tabla_resumen['Cambio_Neto_ha'] = tabla_resumen['2024'] - tabla_resumen['1985']
tabla_resumen['Cambio_Relativo_Porcentaje'] = (tabla_resumen['Cambio_Neto_ha'] / tabla_resumen['1985']) * 100
tabla_resumen['Tasa_Anual_ha_ano'] = tabla_resumen['Cambio_Neto_ha'] / 39 # Período de 39 años entre 1985 y 2024

# 6. Exportar la matriz limpia para las tablas del artículo científico
tabla_resumen.to_csv(archivo_salida, encoding='utf-8')

print("\n📊 MATRIZ DE TRANSFORMACIÓN DEL PAISAJE (LA GUAIRA):")
print(tabla_resumen.round(2).to_string())
print(f"\n✅ Archivo guardado con éxito como: '{archivo_salida}'")
print("\n📝 Cita obligatoria: MapBiomas – Colección 3 de la Serie Anual de Mapas de Cobertura y Uso del Suelo de Venezuela, consultada el 27 de junio de 2026 a través del enlace: https://venezuela.mapbiomas.org")
```
